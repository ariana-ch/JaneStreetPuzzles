# Heapq

## Definition:
<u>heapq</u> data structured for organising lists such that the biggest or smallest element is always at index 0

## The Rule
<u>Rule:</u> (min heap) Every parent node must be smaller or equal to its children nodes. This ensures that the smallest element will always be at the root (0th index).

## Key Functions
`heappush(list, item)`. Adds the item to the list and updates it so that the smallest item is at index 0

`heappop(list)`. Removes and returns the 0th element and updates it so that the smallest item is at index 0

````python
import heapq
import random

my_tasks = [random.randint(0, 10) for _ in range(15)]
print(my_tasks)
heapq.heapify(my_tasks)
print(my_tasks)

heapq.heappush(my_tasks, 20)
heapq.heappush(my_tasks, 5)
heapq.heappush(my_tasks, 12)
print(my_tasks)

print(heapq.heappop(my_tasks))
print(my_tasks)
````

**Example to test with:**

* Current: (0, 0), Lengths: (3, 2) -> Should print (1, 0) and (0, 1).

````python
from typing import Tuple, List
def get_neighbors(current_indices: Tuple[int, ...], list_lengths: Tuple[int, ...]) -> List[Tuple[int, ...]]:
    valid = []
    current_indices_list = list(current_indices)
    for i, idx in enumerate(current_indices):
        length_list_i = list_lengths[i]
        # for idx + 1 to be valid, idx + 1 must be less than length - 1 (zero indexed)
        if idx + 1 < length_list_i:
            valid.append(tuple(current_indices_list[:i] + [idx + 1] + current_indices_list[i+1:]))
    return valid

def print_result(current_indices: Tuple[int, ...], list_lengths: Tuple[int, ...]) -> None:
    result = get_neighbors(current_indices, list_lengths)
    print(f'Current indices: {current_indices}, lengths: {list_lengths} -> {result}')
````

## Radix

Radix means base. It is the number of different digits available in "each place". For example, for decimals the radix is 10 and for binary the radix is 2:
$$
507 = 5 \times 10^2 + 0 \times 10^1 + 7 \times 10^0
$$
and
$$
1011_2 = 1\times 2^3 + 0 \times 2^2 + 1 \times 2^1 + 1 \times 2^0 = 8 + 2 + 1= 11
$$

Any tuple of small digits can be converted into a single integer ID by treating the possible values of each entry of the tuple as its radix. <br>
For example, given a tuple (a, b) with a having 3 possibilities and b having 5 possibilities:

The radix of a is 3 and the radix of b is 5. To encode (a, b):
$$
id = a \times 5 + b
$$
To decode it:
\begin{array}{l}
a = id // 5 \\
b = id \% 5
\end{array}

### Generalisation to length N tuples

Suppose $(x, y, z)$ with ranges:
* $x \in \{0, 1\}$ (radix 2)
* $y \in \{0, 1, 2\}$ (radix 3)
* $z \in \{0, 1, 2, 3\}$ (radix 4)

Then
$$
\text{id} = x \cdot (3\cdot 4) + y \cdot 4 + z
$$
and $z = \text{id} \% 4$, $\text{id}_1 = \text{id} // 4$, $y = \text{id}_1 \% 3$ and $x = \text{id}_1 // 3$.

### Use in Python (tuples, ids, heaps, dictionaries)

It provides a very compact and memory efficient way of mapping tuples/lists to unique integer IDs.

## Mixed Radix System

Let $(r_0, r_1, \ldots, r_{k-1})$ the radices and $(x_0, x_1, \ldots, x_{k-1})$ a tuple of digits. The encoding rule is as follows.

Initialise id = 0
for i in 0:k-1:<br>

    id $\leftarrow$ id * r_i + x_i

To decode.
for i in k-1:0:<br>

    x_i = id % r_i<br>
    id = id // r_i

````python
radix = (4, 3, 2)
def encode(state: Tuple[int, ...], radices: Tuple[int,...] ) -> int:
    state_id = 0
    for i, r in enumerate(radices):
        state_id *= r
        state_id += state[i]
    return state_id
def decode(state_id: int, radices: Tuple[int,...]) -> Tuple[int, ...]:
    state = []
    for i in range(len(radices) - 1, -1, -1):
        radix = radices[i]
        state.insert(0, state_id % radix)
        state_id //= radix
    return tuple(state)

print(encode(state=(3, 2, 1), radices=radix))
print(decode(state_id=17, radices=radix))
````

## Big-Endian Vs Little-Endian
The radix encoding above is called big endian; the most significant and hence dominant number is the one on the left. This is good if we want to preserve the sorting of the tuples.

The Little-Endian system on the other hand (what most computers use), has the most significant number be the last one on the right. This allows for extensibility of the system, i.e. you can increase the length of the tuples without needing to re-calculate all ids.

Example:<br>
Given the tuple $(x_0, \ldots, x_{k-1})$ and the radices $(r_0, \ldots, r_{k-1})$, the Little-Endian encoding is:

$$
\text{id} = x_0 + x_1 \cdot r_0 + x_2 \cdot (r_0 \cdot r_1) + \ldots + r_{k-1} \prod_{i=0}^{k-2} r_i = \sum_{i=0}^{k-1} \prod_{j=-1}^{i-1} r_j, \qquad r_{-1} = 1
$$

and to decode:

$$
\begin{array}{l}
x_{k-1} = \text{id} // \prod_{i=0}^{k-2}\\
x_{k-2} = \left(\text{id} \% \prod_{i=0}^{k-2}r_i\right) // \prod_{j=0}^{k-3} r_j \\
\vdots
\end{array}
$$


````python

def encode_small_endian(state: Tuple[int, ...], radices: Tuple[int,...]) -> int:
    assert len(state) == len(radices)
    assert all([state[i] < r for i, r in enumerate(radices)])
    # construct multipliers
    state_id = state[0]
    m = 1
    for i, rad in enumerate(radices[:-1]):
        m *= rad
        state_id += m * state[i+1]
    return state_id

def decode_small_endian(state_id: int, radices: Tuple[int,...]) -> int:
    multipliers = [math.prod(radices[:i]) for i in range(0, len(radices))]
    state = []
    for i in range(len(radices) - 1, -1, -1):
        state.insert(0, state_id // multipliers[i])
        state_id %= multipliers[i]
    return state
````

````python
from typing import Sequence, Optional, Iterator, Union, Tuple, List
from heapq import heappush, heappop

Number = Union[int, float]


def combos_by_decreasing_sum(
        lists: Sequence[Sequence[Number]],
        *,
        with_sum: bool = False,
        limit: Optional[int] = None,
) -> Iterator[Union[Tuple[float, ...], Tuple[Tuple[float, ...], float]]]:
    """
    Stream combinations choosing one element from each list (in order),
    in monotonically non-increasing order of TOTAL SUM.

    Efficient for "top-k" streaming when the full Cartesian product is huge.
    Uses best-first search on the product lattice with a max-heap.

    Args:
        lists: 1..m sequences, each non-empty.
        with_sum: if True yields (combo, total_sum), else yields combo.
        limit: optional cap on number of yielded combos.

    Yields:
        combo tuple (floats), or (combo tuple, total_sum) if with_sum=True.
    """
    m = len(lists)
    if m == 0:
        return
        yield  # pragma: no cover

    # No combos if any list is empty
    for lst in lists:
        if not lst:
            return

    # Sort each list descending so index 0 is the max contribution.
    arr: List[List[float]] = [sorted((float(x) for x in lst), reverse=True) for lst in lists]
    lens = [len(a) for a in arr]

    # Mixed-radix multipliers so we can encode an index-vector into one integer id.
    # id = sum(idx[d] * mult[d]) with mult[d] = product_{i<d} lens[i]
    mult = [1] * m
    for d in range(1, m):
        mult[d] = mult[d - 1] * lens[d - 1]

    def decode(id_: int) -> List[int]:
        idx = [0] * m
        for d in range(m):
            idx[d] = (id_ // mult[d]) % lens[d]
        return idx

    start_id = 0  # all indices = 0
    start_sum = 0.0
    for d in range(m):
        start_sum += arr[d][0]

    # Heap stores (-sum, state_id). visited stores state_id ints (much smaller than tuples).
    heap: List[Tuple[float, int]] = [(-start_sum, start_id)]
    visited = {start_id}

    yielded = 0
    while heap:
        neg_s, state_id = heappop(heap)
        s = -neg_s

        idx = decode(state_id)
        combo = tuple(arr[d][idx[d]] for d in range(m))

        if with_sum:
            yield combo, s
        else:
            yield combo

        yielded += 1
        if limit is not None and yielded >= limit:
            return

        # Generate neighbours by increasing one coordinate by 1.
        # Because lists are sorted descending, this never increases the sum,
        # so best-first search yields non-increasing sums globally.
        for d in range(m):
            i = idx[d]
            j = i + 1
            if j >= lens[d]:
                continue

            new_id = state_id + mult[d]  # increment digit d in mixed radix representation
            if new_id in visited:
                continue
            visited.add(new_id)

            new_sum = s - arr[d][i] + arr[d][j]
            heappush(heap, (-new_sum, new_id))
````

## Solve Jane Street Puzzle

````python

from typing import Sequence, Any
import heapq

OPTIONS = dict(((i, j), list(range(9, -1, -1))) for i in range(5) for j in range(5))
FIXED = {(1, 4): [8, 6, 4, 2, 0],
         (3, 4): [8, 6, 4, 2, 0],
         (4, 0): [8, 6, 4, 2, 0],
         (4, 2): [8, 6, 4, 2, 0]}
OPTIONS.update(FIXED)

def is_valid(arr: Sequence[int]) -> bool:
    row_len = int(math.sqrt(len(arr)))
    square = [arr[i*row_len:(i+1)*row_len] for i in range(row_len)]
    m = [10**i for i in range(row_len-1, -1, -1)]
    # check rows
    for i in range(row_len):
        if not sum([square[i][j] * k for j, k in enumerate(m)]) % (i + 1):
            return False
    for i in range(row_len):
        if not sum([square[j][i] * k for j, k in enumerate(m)]) % (i + 6):
            return False
    return True

def ordered_combinations(options: Sequence[Sequence[int]]) -> Any:
    options = [sorted(x, reverse=True) for x in options]
    lens = [len(x) for x in options]
    strides = [math.prod(lens[:i]) for i in range(len(lens))]
    def decode(id_: int) -> List[int]:
        state = []
        for m in range(len(strides)-1, -1, -1):
            state.insert(0, (id_ // strides[m]))
            id_ %= strides[m]
        return state
    start_id = 0
    current_state = decode(start_id)
    start_sum = sum([options[i][j] for i, j in enumerate(current_state)])
    max_id = math.prod(lens) - 1
    heap = [(-start_sum, start_id)]
    visited: set = set()

    while heap:
        neg_s, state_id = heapq.heappop(heap)
        current_state = decode(state_id)
        current_array = [options[i][j] for i, j in enumerate(current_state)]
        yield current_array

        # find the neighbors and add them to the heap
        for i, d in enumerate(strides):
            new_id = state_id + d
            if new_id in visited or new_id > max_id:
                continue
            visited.add(new_id)
            remove_entry = options[i][current_state[i]]
            add_entry = options[i][current_state[i] + 1]
            new_neg_s = neg_s + remove_entry - add_entry
            heapq.heappush(heap, (new_neg_s, new_id))

def search():
    found = False
    combinations = ordered_combinations(list(OPTIONS.values()))
    combination = None
    while not found:
        combination = next(combinations)
        found = is_valid(combination)
    return combination
result = search()
````
